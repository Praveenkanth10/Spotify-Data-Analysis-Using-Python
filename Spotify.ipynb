```python
most_popular=data.query('popularity>90',inplace=False).sort_values('popularity',ascending=False)
most_popular[:10]
```

```python
# Check if 'release_date' is already the index
if 'release_date' not in data.columns:
    # If not, try setting it as the index
    try:
        data.set_index("release_date", inplace=True)
    except KeyError:
        print("Column 'release_date' not found in DataFrame. It might already be the index.")
else:
    print("'release_date' is already the index.")

# Convert index to datetime
data.index = pd.to_datetime(data.index, errors='coerce', format='%Y-%m-%d', infer_datetime_format=True)

data.head()
```

```python
data[["artists"]].iloc[100]
```

```python
# Check if 'duration_ms' column exists before proceeding
if 'duration_ms' in data.columns:
    data["duration"]=data["duration_ms"].apply(lambda x: round(x/1000))
    data.drop("duration_ms",inplace=True,axis=1)

    print("Column 'duration' created and 'duration_ms' dropped successfully.") # Optional print for confirmation
else:
    print("Column 'duration_ms' not found. It might have already been processed.")
```

```python
data.duration.head()
```

```python
numerical_features = data.select_dtypes(include=np.number).columns
corr_df = data[numerical_features].drop(["key", "mode", "explicit"], axis=1, errors='ignore').corr(method="pearson")
plt.figure(figsize=(14, 6))
heatmap = sns.heatmap(corr_df, annot=True, fmt=".1g", vmin=-1, vmax=1, center=0, cmap="inferno", linewidths=1, linecolor="Black")
heatmap.set_title("Correlation Heatmap Between Variables")
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation=90)
"plt.show()
```
